#"Cheese Kills: The Curious Case of Cheese Consumption and Death by Bedsheet"

#The Two Variables and Data Sources

##1. Per Capita Cheese Consumption in the U.S. (lbs/year)

Source: U.S. Department of Agriculture (USDA)

[Data compiled by Statista USDA ERS](https://www.ers.usda.gov/data-products/dairy-data/)

##2. Number of People Who Died by Becoming Tangled in Their Bedsheets (U.S.)

Source: U.S. Centers for Disease Control and Prevention (CDC), National Vital Statistics

[Available through the CDC at National Vital Statistics Reports (NVSR) and their mortality data files.](https://www.cdc.gov/nchs/products/nvsr.htm)

#Evidence of Correlation
A Pearson correlation coefficient between the two variables (2000–2009):

r ≈ 0.95, which is extremely high for real-world data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

#dataset
data = {
    "Year": list(range(2000, 2010)),
    "Cheese_Consumption_lbs": [29.8, 30.1, 30.5, 30.6, 31.3, 31.7, 32.6, 32.8, 32.6, 32.8],
    "Bedsheet_Deaths": [327, 456, 509, 497, 518, 562, 573, 539, 566, 617]
}
df = pd.DataFrame(data)

#CSV
df.to_csv("cheese_vs_bedsheet_deaths.csv", index=False)

#correlation
correlation, p_value = pearsonr(df["Cheese_Consumption_lbs"], df["Bedsheet_Deaths"])
print(f"Correlation: {correlation:.2f}, p-value: {p_value:.4f}")

#plot
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
ax = sns.lineplot(x="Year", y="Cheese_Consumption_lbs", data=df, label="Cheese Consumption (lbs/person)", marker="o")
ax2 = ax.twinx()
sns.lineplot(x="Year", y="Bedsheet_Deaths", data=df, label="Bedsheet Deaths", ax=ax2, color="r", marker="o")
ax.set_ylabel("Cheese Consumption (lbs/person)")
ax2.set_ylabel("Bedsheet Deaths")
plt.title("Correlation Between Cheese Consumption and Bedsheet Deaths (2000–2009)")
ax.legend(loc="upper left")
ax2.legend(loc="upper right")
plt.tight_layout()
plt.savefig("cheese_vs_bedsheet_plot.png")
plt.show()
```

A dual-axis line graph showing:

X-axis: Years (2000–2009)

Left Y-axis: Cheese Consumption (lbs/person)

Right Y-axis: Deaths by Bedsheet Tangling

Both lines rise and fall nearly identically

Both cheese consumption and bedsheet entanglement deaths follow a near-identical trend over a decade. Despite appearing unrelated, the visual similarity and strong statistical correlation suggest some shared underlying pattern—or at least a fun narrative worth inventing.

#Explanation (The Why)
The answer is surprisingly obvious once you connect the dots:
Cheese makes you sleepy. Sleep is dangerous.

Here's the chain of events:

Increased cheese consumption leads to an increase in tryptophan intake (an amino acid that promotes sleepiness).

People eat lots of cheese and fall asleep faster and deeper.

In a deep sleep state, individuals are less aware of their surroundings, less responsive to discomfort, and possibly more tangled in sheets.

Result? A tragic uptick in fatal sheet entrapments.

In other words: The more cheese you eat, the more likely your sheets will kill you.
